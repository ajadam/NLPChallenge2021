```python
import pandas as pd
```

```python
def macro_disparate_impact(people):
    counts = people.groupby(['job', 'gender']).size().unstack('gender')
    counts['disparate_impact'] = counts[['M', 'F']].max(axis='columns') / counts[['M', 'F']].min(axis='columns')
    return counts['disparate_impact'].mean()
```

```python
PATH = '/Users/antoineadam/git/NLPChallenge2021/data/'
genders = pd.read_json(PATH+'test.json').set_index('Id')['gender']
genders.head()
```

```python
names = pd.read_csv(PATH+'categories_string.csv')['0'].to_dict()
jobs = pd.read_csv('/Users/antoineadam/git/NLPChallenge2021/code/models/ensemble_v4/distilBERT.csv', index_col='Id')['Category']
jobs = jobs.map(names)
jobs = jobs.rename('job')
jobs.head()
```

```python
people = pd.concat((jobs, genders), axis='columns')
people.head()
```

```python
macro_disparate_impact(people)
```
